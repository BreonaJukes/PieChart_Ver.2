```python
import matplotlib.pyplot as plt

#this is where the data from the file will go
slices = []
activities = []

#this reads the data from the file then splits it into a list then appends this list to
#the empty activities list
for line in open('Data.txt', 'r'):
    Data = [i for i in line.split()]
    activities.append(Data[0])
    
#this reads the data from the file then splits it into a list then appends this list to
#the empty slices list
    New_Data = [j for j in Data[1].split()]
    slices.append(Data[1])

#this assigns the colors to the slices
cols = ['c', 'm', 'r', 'b']

#this plots the pie chart
plt.pie(slices,
       labels = activities,
       colors = cols,
       startangle = 90,
       shadow = True,
       explode = (0, 0, 0, 0),
       autopct = '%1.1f%%')

#this displays the pie chart
plt.title('Pie Chart')
plt.show()
```
